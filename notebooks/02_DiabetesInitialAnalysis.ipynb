## Summary Stats for Age 

```python
df['Age_of_Participant'].describe()
```

```python
def age_group(age):
    if age >= 16 and age <= 35: 
        return "16-35"
    elif age >= 36 and age <= 50: 
        return "36-50"
    elif age >= 51 and age <= 66: 
        return "51-65"
    else: 
        return "66 and above"
```

```python
df['Age_Group'] = df['Age_of_Participant'].apply(age_group)
df.head() 
```

```python
# check age group 
df[['Age_of_Participant', 'Age_Group']].head() 
```
